```python
# IMPORT DATA
import pandas as pd
df = pd.read_excel('Superstore Dataset.xlsx')
df.head(3)
```

```python
# UBAH TIPE DATA SALES DARI STRING JADI FLOAT
df['Sales'] = df['Sales'].str.replace('$', '')
df['Sales'] = df['Sales'].str.replace(',', '')
df['Sales'] = df['Sales'].astype(float)
df
```

HANDLING MISSING VALUES

```python
# PERIKSA KOLOM YANG MENGANDUNG MISSING VALUES
jumlah_kosong = df.isnull().sum()
print(jumlah_kosong)

# kolom_kosong = jumlah_kosong[jumlah_kosong > 0]
# print(kolom_kosong)
```

```python
# HANDLING MISSING VALUE
# Customer Name (isi based on customer ID, kalau gaada diisi unknown)
referensi_nama = df[['Customer ID', 'Customer Name']].dropna().drop_duplicates('Customer ID')
referensi_nama = referensi_nama.set_index('Customer ID')['Customer Name']
df['Customer Name'] = df['Customer Name'].fillna(df['Customer ID'].map(referensi_nama)).fillna('Unknown')

# Postal Code (Logic Mapping Sederhana)
referensi_pos = df[['City', 'Postal Code']].dropna().drop_duplicates('City')
referensi_pos = referensi_pos.set_index('City')['Postal Code'] # membuat master postal code
df['Postal Code'] = df['Postal Code'].fillna(df['City'].map(referensi_pos))

# Postal Code: Mengubah tipe data menjadi string
df['Postal Code'] = df['Postal Code'].astype(str)

# Discount (MAR di Henderson) -- isi dengan mean discount
rata_discount_henderson = df[df['City'] == 'Henderson']['Discount'].mean()
df['Discount'] = df['Discount'].fillna(rata_discount_henderson)

# Langsung isi dengan angka 0
df['Discount'] = df['Discount'].fillna(0)
```

```python
# Cek kembali apakah masih ada missing value
jumlah_kosong = df.isnull().sum()
kolom_kosong = jumlah_kosong[jumlah_kosong > 0]
print(kolom_kosong)

data_sisa_kosong = df[df['Postal Code'].isnull()]
print(data_sisa_kosong)
```

```python
# Hapus Data Postal Code yang Missing (Opsional tergantung tujuan analisa)
df = df.dropna(subset=['Postal Code'])
```

HANDLING DUPLICATIONS

```python
# CEK DUPLIKASI BARIS
jumlah_duplikat = df.duplicated().sum()
print(jumlah_duplikat)
```

```python
# Handling Duplicates (Duplikasi Baris)
df = df.drop_duplicates()
```

```python
# Mencari Customer ID yang memiliki lebih dari 1 nama unik
df_partial = df.groupby('Customer ID')['Customer Name'].unique().reset_index()
df_partial['Jumlah_Nama'] = df_partial['Customer Name'].apply(len)

# Tampilkan hanya yang bermasalah
id_bermasalah = df_partial[df_partial['Jumlah_Nama'] > 1]
print(id_bermasalah)
```

```python
# Handling Duplikat Parsial
# Urutkan berdasarkan tanggal terbaru dulu agar nama terakhir yang diambil benar
df = df.sort_values(by='Order Date', ascending=True)
df['Customer Name'] = df.groupby('Customer ID')['Customer Name'].transform('last')
```

```python
# CEK DUPLIKASI KOLOM
kolom_duplikat = []
for i in range(len(df.columns)):
    for j in range(i + 1, len(df.columns)):
        kolom1 = df.columns[i]
        kolom2 = df.columns[j]
        # Jika seluruh isi kolom sama persis
        if df[kolom1].equals(df[kolom2]):
            kolom_duplikat.append(f"{kolom1} sama dengan {kolom2}")
print(kolom_duplikat)
```

```python
# CEK DUPLIKASI KOLOM STATE
kolom_a = 'State'
kolom_b = 'State_Code'

# Kelompokkan berdasarkan State, lalu hitung ada berapa State_Code di tiap State
cek_hubungan = df.groupby(kolom_a)[kolom_b].nunique()

# Mencari tahu mana yang hubungannya "selingkuh" (lebih dari 1 pasangan)
masalah = cek_hubungan[cek_hubungan > 1]
print(masalah)
```

```python
# HANDLING DUPLIKASI KOLOM
df = df.drop(columns=['Transaction_ID'])
df = df.drop(columns=['State_Code'])
```

INCONSISTENCY FORMAT

```python
# INKONSISTENSI FORMAT KOLOM STATE & SEGMENT
print("Variasi tulisan di kolom 'Segment':")
print(df['Segment'].unique())

print("\nVariasi tulisan di kolom 'State':")
print(df['State'].unique())
```

```python
# Segment: Pakai .str.title() untuk otomatis mengubah "CONSUMER" atau "consumer" jadi "Consumer"
# 'Consumer' 'CONSUMER' 'consumer' 'Consumerrr'
df['Segment'] = df['Segment'].str.title()
# Memperbaiki typo spesifik
df['Segment'] = df['Segment'].replace('Consumerrr', 'Consumer')

# Abbreviation State
df['State'] = df['State'].replace('CA', 'California')
df['State'] = df['State'].replace('Calif.', 'California')
```

HANDLING OUTLIERS

```python
# MEMBERSIHKAN OUTLIER SALES
q1_sales = df['Sales'].quantile(0.25)
q3_sales = df['Sales'].quantile(0.75)
iqr_sales = q3_sales - q1_sales
batas_bawah_sales = q1_sales - (1.5 * iqr_sales)
batas_atas_sales = q3_sales + (1.5 * iqr_sales)
```

```python
# Melihat nilai outlier
outlier_sales = df[(df['Sales'] < batas_bawah_sales) | (df['Sales'] > batas_atas_sales)]
print(outlier_sales)
```

```python
# Data Sales tanpa Outlier
df = df[(df['Sales'] >= batas_bawah_sales) & (df['Sales'] <= batas_atas_sales)]
```

```python
# MEMBERSIHKAN OUTLIER QUANTITY
q1_qty = df['Quantity'].quantile(0.25)
q3_qty = df['Quantity'].quantile(0.75)
iqr_qty = q3_qty - q1_qty
batas_bawah_qty = q1_qty - (1.5 * iqr_qty)
batas_atas_qty = q3_qty + (1.5 * iqr_qty)

outlier_qty = df[(df['Quantity'] < batas_bawah_qty) | (df['Quantity'] > batas_atas_qty)]
print(outlier_qty)

# df = df[(df['Quantity'] >= batas_bawah_qty) & (df['Quantity'] <= batas_atas_qty)]
```

```python
# MEMBERSIHKAN OUTLIER PROFIT
q1_profit = df['Profit'].quantile(0.25)
q3_profit = df['Profit'].quantile(0.75)
iqr_profit = q3_profit - q1_profit
batas_bawah_profit = q1_profit - (1.5 * iqr_profit)
batas_atas_profit = q3_profit + (1.5 * iqr_profit)

outlier_profit = df[(df['Profit'] < batas_bawah_profit) | (df['Profit'] > batas_atas_profit)]
print(outlier_profit)

# df = df[(df['Profit'] >= batas_bawah_profit) & (df['Profit'] <= batas_atas_profit)]
```

DATA BINNING

```python
# MEMBAGI SALES MENJADI 3 KATEGORI
df['Performance_Level'] = pd.qcut(df['Sales'], q=3, labels=["Low", "Medium", "High"])
df
```

```python
# BINNING SHIPPING
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Menghitung selisih hari dalam bentuk angka
df['Days_to_Ship'] = (df['Ship Date'] - df['Order Date']).dt.days

batas = [-1, 2, 5, 999] # -1 sebagai batas bawah karena dihitung mulai 0 hari, 999 angka terbesar
nama_label = ["Fast", "Normal", "Late"]
df['Shipping_Status'] = pd.cut(df['Days_to_Ship'], bins=batas, labels=nama_label)
df
```

UNIVARIATE ANALYSIS

```python
# Mengetahui Mean & Median Sales
sales_stats = df['Sales'].describe()
mean_val = sales_stats['mean']
median_val = df['Sales'].median()

print(sales_stats)
print(median_val)
```

```python
# Deteksi Skewness dari nilai Mean dan Median
skew_val = df['Sales'].skew() # Fungsi bawaan pandas untuk hitung kemiringan

print(f"Hasil Analisis Kolom Sales:")
print(f"Mean   : {mean_val:.2f}")
print(f"Median : {median_val:.2f}")
print(f"Nilai Skewness: {skew_val:.2f}")

# Logika Deteksi
selisih_persen = abs(mean_val - median_val) / median_val * 100

if skew_val > 1:
    print("Status: POSITIVE SKEW (Miring Kanan)")
    print("Artinya: Ada transaksi sangat besar yang menarik rata-rata ke atas.")
elif skew_val < -1:
    print("Status: NEGATIVE SKEW (Miring Kiri)")
    print("Artinya: Ada transaksi yang nilainya sangat rendah/anjlok dibanding biasanya.")
else:
    print("Status: NORMAL / SIMETRIS")
    print("Artinya: Distribusi data cukup sehat, rata-rata bisa dipercaya.")

if selisih_persen > 20:
    print(f"Peringatan: Selisih Mean & Median cukup besar ({selisih_persen:.1f}%).")
    print("Saran: Gunakan MEDIAN sebagai patokan performa harian daripada Mean.")
```

```python
# Analisis Profit
mean_profit = df['Profit'].mean()
std_profit = df['Profit'].std()
min_profit = df['Profit'].min()
max_profit = df['Profit'].max()

# Hitung Coefficient of Variation (CV)
# CV adalah standar deviasi dibagi rata-rata.
# Jika CV > 1 (atau > 100%), artinya data sangat fluktuatif.
cv_profit = abs(std_profit / mean_profit)

print(f"Rata-rata Profit    : {mean_profit:.2f}")
print(f"Standar Deviasi     : {std_profit:.2f}")
print(f"Rentang (Min ke Max): {min_profit:.2f} s/d {max_profit:.2f}")
print(f"Koefisien Variasi   : {cv_profit:.2f}")

# Logika Pengambilan Keputusan
if cv_profit > 1:
    status = "SANGAT FLUKTUATIF"
    penjelasan = ("Nilai standar deviasi lebih besar dari rata-ratanya. "
                  "Hal ini artinya jarak antar keuntungan transaksi satu dengan lainnya sangat jauh berbeda.")
else:
    status = "CUKUP STABIL"
    penjelasan = "Nilai standar deviasi kecil, artinya profit tiap transaksi cenderung mirip dengan rata-ratanya."

print(f"Status Profit: {status}")
print(f"Penjelasan   : {penjelasan}")

# Tambahan: Cek adanya kerugian (Loss)
if min_profit < 0:
    print(f"Peringatan: Ditemukan kerugian maksimal hingga {abs(min_profit):.2f}. "
          "Hal ini yang menyebabkan angka fluktuasi menjadi tinggi.")
```
